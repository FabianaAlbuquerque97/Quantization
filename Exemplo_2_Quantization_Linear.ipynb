```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Função para quantização linear
def quantize_linear(data, levels):

  # Define o intervalo de quantização
  min_value = np.min(data)
  max_value = np.max(data)
  interval = (max_value - min_value) / levels

  # Quantiza os dados
  quantized_data = np.round((data - min_value) / interval) * interval + min_value

  return quantized_data

# Gera dados de exemplo
data = np.random.rand(100) * 10  # Dados aleatórios entre 0 e 10

# Define o número de níveis de quantização
levels = 5

# Quantiza os dados
quantized_data = quantize_linear(data, levels)
```

```python
# Plota os dados originais e quantizados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(data, label="Dados originais")
plt.title("Dados originais")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(quantized_data, label="Dados quantizados")
plt.title(f"Dados quantizados ({levels} níveis)")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()

plt.tight_layout()
plt.show()
```
